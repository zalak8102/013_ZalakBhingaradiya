```python
#Import scikit-learn dataset library
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
import numpy as np

#Load dataset
wine = datasets.load_wine()
```

```python
#import the necessary module
from sklearn.model_selection import train_test_split
#split data set into train and test sets
data_train, data_test, target_train, target_test = train_test_split(wine.data, wine.target, test_size = 0.20, random_state = 13)
```

```python

#Create a Decision Tree Classifier (using Gini)
model = DecisionTreeClassifier(criterion='gini', max_leaf_nodes=13, random_state=13)
#Train the model using the training sets
model.fit(data_train, target_train)
```

```python
# Predict the classes of test data
target_predicated = model.predict(data_test)
```

```python
# Model Accuracy, how often is the classifier correct?
from sklearn import metrics

print("Accuracy is :- ",metrics.accuracy_score(target_test, target_predicated))

# print precision and recall

precision = metrics.precision_score(target_test, target_predicated, average='macro')
recall = metrics.recall_score(target_test, target_predicated, average='macro')

print("precision :- ",precision)
print("recall :- ",recall)
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
tree.plot_tree(model,feature_names=wine.feature_names,
               class_names=wine.target_names,
               filled=True)
plt.show()
```
